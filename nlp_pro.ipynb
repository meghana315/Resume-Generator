```python
from transformers import pipeline
generator = pipeline('text-generation', model='gpt2')

def generate_social_media_content(prompt, max_length=50, num_return_sequences=1):
    generated_texts = generator(prompt, max_length=max_length, num_return_sequences=num_return_sequences)
    return [text['generated_text'] for text in generated_texts]

prompt = input("Enter a prompt for social media content: ")

max_length = int(input("Enter the maximum length of the generated content (e.g., 50): "))

num_return_sequences = int(input("Enter the number of content variations to generate (e.g., 3): "))

generated_content = generate_social_media_content(prompt, max_length=max_length, num_return_sequences=num_return_sequences)

for idx, content in enumerate(generated_content):
    print(f"Generated Content {idx + 1}:\n{content}\n")
```
